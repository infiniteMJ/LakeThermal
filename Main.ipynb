## Background

There are several aspects of thermocline that we consider, such as magnitude , depth and thickness. In this project, two of such parameters, temperature shear and thermocline lower boundary, are considered in the model and evaluation below. Thus, we first introduce the definition of such parameter and why it is important and worthy to consider. For better understanding, we showed a temperature profile from observation and two different model predictions for the same day.

Screen Shot 2022-03-08 at 3.59.07 PM.png

First of all, temperature shear is the temperature difference between near-surface and near-basin. This temperature shear shows the magnitude of lake stratification. The larger temperature shear means more stratified lake if other conditions are same. We considered difference between model prediction and observation as error and used that quantity for evaluation. Furthermore, we made a new loss function which consider same quantity as another loss additional to RMSE and energy loss. The exact calculation and the result is given below and will be analyzed further. Figure 2 shows that RMSE does not always covary with temperature shear and this emphasize the need of another loss function.

Figure2

The second parameter is depth of lower boundary of thermocline (or metalimnion). It is defined using vertical temperature gradient, which is interpolated for more accurate comparison. This parameter tells us whether the thermocline can be easily removed or not. If the lower boundary is too shallow, surface wind can mix the lake deep enough to break the stratification. Although, typical wind cannot entirely removes it, reducing the stratification can make disturbaces easier to penetrate through thermocline.

## Our Additional Loss Function 
In addition to the model loss function, we consider an additional loss `calculate_stratification` which takes advantage of the temperature shear, as mentioned in the first section. Because of the sparse observation data, we used averaged value of 0 ~ 3m and 17 ~ 20m for near-surface and near-basin, respectively. For the solidity of the loss function, we considered temperature shear as loss for the lake profile that are enough stratified, which we assumed here to have temperature shear over 5K in observation.
$$
\mathcal{L}(\hat{T}, T)=\sum_{i=1}^n |(T)_i-(\hat{T})_i|\mathbb{1}_{\{(T)_i\geq 5\}}
$$

```python
#@title
def transformTempToDensity(temp):
    # converts temperature to density
    # parameter:
        # @temp: single value or array of temperatures to be transformed
    densities = 1000 * (1 - ((temp + 288.9414) * (temp - 3.9863)**2) / (508929.2 * (temp + 68.12963)))
    return densities

def calculate_lake_energy(temps, densities, depth_areas):
    # calculate the total energy of the lake for every timestep
    # sum over all layers the (depth cross-sectional area)*temp*density*layer_height)
    # then multiply by the specific heat of water 
    dz = 0.5 # thickness for each layer, hardcoded for now
    cw = 4186 # specific heat of water
    depth_areas = torch.reshape(depth_areas, (-1, 1))
    energy = torch.sum(depth_areas * temps * densities * dz * cw, axis=0)
    return energy

def calculate_lake_energy_deltas(energies, combine_days, surface_area):
    # given a time series of energies, compute and return the differences
    # between each time step, or time step interval (parameter @combine_days)
    # as specified by parameter @combine_days
    time = 86400 #seconds per day
    energy_deltas = (energies[1:] - energies[:-1]) / (time * surface_area)
    return energy_deltas

def calculate_vapour_pressure_saturated(temp):
    # returns in miilibars
    # Converted pow function to exp function workaround pytorch not having autograd implemented for pow
    exponent = (9.28603523 - (2332.37885 / (temp + 273.15))) * np.log(10)
    return torch.exp(exponent)

def calculate_vapour_pressure_air(rel_hum, temp):
    rh_scaling_factor = 1
    return rh_scaling_factor * (rel_hum / 100) * calculate_vapour_pressure_saturated(temp)


def calculate_wind_speed_10m(ws, ref_height = 2.):
    # from GLM code glm_surface.c
    c_z0 = torch.tensor(0.001) #default roughness
    return ws * (torch.log(10.0 / c_z0) / torch.log(ref_height / c_z0))

def calculate_air_density(air_temp, rh):
    # returns air density in kg / m^3
    # equation from page 13 GLM/GLEON paper(et al Hipsey)
    # Ratio of the molecular (or molar) weight of water to dry air
    mwrw2a = 18.016 / 28.966
    c_gas = 1.0e3 * 8.31436 / 28.966

    # atmospheric pressure
    p = 1013. #mb

    # water vapor pressure
    vapPressure = calculate_vapour_pressure_air(rh, air_temp)

    # water vapor mixing ratio (from GLM code glm_surface.c)
    r = mwrw2a * vapPressure / (p - vapPressure)
    return (1.0 / c_gas * (1 + r)/(1 + r / mwrw2a) * p / (air_temp + 273.15)) * 100

def calculate_heat_flux_sensible(surf_temp, air_temp, rel_hum, wind_speed):
    # equation 22 in GLM/GLEON paper(et al Hipsey)
    # GLM code ->  Q_sensibleheat = -CH * (rho_air * 1005.) * WindSp * (Lake[surfLayer].Temp - MetData.AirTemp);
    # calculate air density 
    rho_a = calculate_air_density(air_temp, rel_hum)

    # specific heat capacity of air in J/(kg*C)
    c_a = 1005.

    # bulk aerodynamic coefficient for sensible heat transfer
    c_H = 0.0013

    # wind speed at 10m
    U_10 = calculate_wind_speed_10m(wind_speed)
    return -rho_a * c_a * c_H * U_10 * (surf_temp - air_temp)

def calculate_heat_flux_latent(surf_temp, air_temp, rel_hum, wind_speed):
    # equation 23 in GLM/GLEON paper(et al Hipsey)
    # GLM code-> Q_latentheat = -CE * rho_air * Latent_Heat_Evap * (0.622/p_atm) * WindSp * (SatVap_surface - MetData.SatVapDef)
    # where,         SatVap_surface = saturated_vapour(Lake[surfLayer].Temp);
    #                rho_air = atm_density(p_atm*100.0,MetData.SatVapDef,MetData.AirTemp);
    # air density in kg/m^3
    rho_a = calculate_air_density(air_temp, rel_hum)

    # bulk aerodynamic coefficient for latent heat transfer
    c_E = 0.0013

    # latent heat of vaporization (J/kg)
    lambda_v = 2.453e6

    # wind speed at 10m height
    # U_10 = wind_speed
    U_10 = calculate_wind_speed_10m(wind_speed)
    # 
    # ratio of molecular weight of water to that of dry air
    omega = 0.622

    # air pressure in mb
    p = 1013.

    e_s = calculate_vapour_pressure_saturated(surf_temp)
    e_a = calculate_vapour_pressure_air(rel_hum, air_temp)
    return -rho_a * c_E * lambda_v * U_10 * (omega / p) * (e_s - e_a)

def calculate_energy_fluxes(phys, surf_temps, combine_days):    
    e_s = 0.985 # emissivity of water, given by Jordan
    alpha_sw = 0.07 # shortwave albedo, given by Jordan Read
    alpha_lw = 0.03 # longwave, albeda, given by Jordan Read
    sigma = 5.67e-8 # Stefan-Baltzmann constant
    R_sw_arr = phys[:-1,2] + (phys[1:,2] - phys[:-1,2]) / 2
    R_lw_arr = phys[:-1,3] + (phys[1:,3] - phys[:-1,3]) / 2
    R_lw_out_arr = e_s * sigma * (torch.pow(surf_temps[:] + 273.15, 4))
    R_lw_out_arr = R_lw_out_arr[:-1] + (R_lw_out_arr[1:] - R_lw_out_arr[:-1]) / 2

    air_temp = phys[:-1,4] 
    air_temp2 = phys[1:,4]
    rel_hum = phys[:-1,5]
    rel_hum2 = phys[1:,5]
    ws = phys[:-1, 6]
    ws2 = phys[1:,6]
    t_s = surf_temps[:-1]
    t_s2 = surf_temps[1:]
    E = calculate_heat_flux_latent(t_s, air_temp, rel_hum, ws)
    H = calculate_heat_flux_sensible(t_s, air_temp, rel_hum, ws)
    E2 = calculate_heat_flux_latent(t_s2, air_temp2, rel_hum2, ws2)
    H2 = calculate_heat_flux_sensible(t_s2, air_temp2, rel_hum2, ws2)
    E = (E + E2) / 2
    H = (H + H2) / 2
    fluxes = (R_sw_arr[:-1] * (1-alpha_sw) + R_lw_arr[:-1] * (1-alpha_lw) - R_lw_out_arr[:-1] + E[:-1] + H[:-1])
    return fluxes

def calculate_ec_loss(inputs, outputs, phys, depth_areas, n_depths, ec_threshold, combine_days=1):
    # description: calculates energy conservation loss
    # parameters: 
        # @inputs: features
        # @outputs: labels
        # @phys: features(not standardized) of sw_radiation, lw_radiation, etc
        # @ labels modeled temp (will not used in loss, only for test) !!! DO NOT MODIFY THIS LINE !!!
        # @depth_areas: cross-sectional area of each depth
        # @n_depths: number of depths
        # @use_gpu: gpu flag
        # @combine_days: how many days to look back to see if energy is conserved
    #*********************************************************************************
    n_sets = int(inputs.shape[0] / n_depths)
    densities = transformTempToDensity(outputs)
    diff_per_set_r = torch.empty(n_sets) 
    for i in range(n_sets):
        # loop through sets of n_depths
        # indices
        start_index = i * n_depths
        end_index = (i + 1) * n_depths
        # calculate lake energy for each timestep
        lake_energies = calculate_lake_energy(outputs[start_index:end_index, :], densities[start_index:end_index, :], depth_areas)
        # calculate energy change in each timestep
        lake_energy_deltas = calculate_lake_energy_deltas(lake_energies, combine_days, depth_areas[0])
        lake_energy_deltas = lake_energy_deltas[1:]
        # calculate sum of energy flux into or out of the lake at each timestep
        lake_energy_fluxes = calculate_energy_fluxes(phys[start_index, :, :], outputs[start_index, :], combine_days)
        ### can use this to plot energy delta and flux over time to see if they line up
        diff_vec = torch.abs(lake_energy_deltas - lake_energy_fluxes) 
        tmp_mask = 1 - phys[start_index+1, 1:-1, 9] 
        tmp_loss = torch.mean(diff_vec * tmp_mask)
        diff_per_set_r[i] = tmp_loss 
    diff_per_set = torch.clamp(diff_per_set_r - ec_threshold, min=0, max=999999)
    return torch.mean(diff_per_set), diff_vec, diff_per_set_r, diff_per_set 

def calculate_stratification(outputs_t, n_depths, target_t):
  n_sets = int(outputs_t.shape[0] / n_depths)
  outputs = outputs_t.cpu().detach().numpy()
  target = target_t.cpu().detach().numpy()
  target[target==0] = np.nan
  dT_pred = np.zeros((n_sets,outputs.shape[1]))
  dT_target = np.zeros((n_sets,outputs.shape[1]))
  loss = np.zeros((n_sets,outputs.shape[1]))
  for i in range(n_sets):
    start_index = i * n_depths
    end_index = (i + 1) * n_depths
    dT_pred[i,:] = np.mean(outputs[start_index:start_index+6,:],axis=0)-np.mean(outputs[end_index-15:end_index-9,:],axis=0)
    dT_target[i,:] = np.nanmean(target[start_index:start_index+6,:],axis=0)-np.nanmean(target[end_index-15:end_index-9,:],axis=0)
    for j in range(outputs.shape[1]):
      if (dT_target[i,j] < 5) & (np.isnan(dT_target[i,j])):
        loss[i,j] = 0
      else:
        loss[i,j] = np.abs(dT_target[i,j]-dT_pred[i,j])
  return torch.tensor(np.nanmean(loss)).to(device)
```

## LSTM model structures
We set up a relatively standard LSTM network. See [the PyTorch tutorial on LSTM](https://pytorch.org/docs/stable/generated/torch.nn.LSTM.html) for details. 

There are two LSTM structures `LSTMnetModified` and `LSTMnetOrigin`. The difference between the two structures is that we add two `tanh(x)` activations in the input and output of the LSTM layer. We believe that the two additional activations could help smooth the gradient and avoid gradient vanish problem.

#### modified version

```python
#@title
class LSTMnetModified(nn.Module):
    def __init__(self, input_dim, output_dim, hidden_dim, n_layers):
        super(LSTMnetModified, self).__init__()
        self.output_size = output_dim
        self.n_layers = n_layers
        self.hidden_dim = hidden_dim
        self.lstm = nn.LSTM(input_size = input_dim, 
                            hidden_size = hidden_dim, 
                            num_layers = n_layers, 
                            batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)
        self.tanh = nn.Tanh()
        self.leakyRelu = nn.LeakyReLU()
        
    def forward(self, x):
        batch_size = x.size(0)
        x = self.tanh(x)
        lstm_out, hidden = self.lstm(x)
        lstm_out = lstm_out.contiguous().view(-1, self.hidden_dim)
        lstm_out = self.tanh(lstm_out)
        out = self.fc(lstm_out)
        out = out.view(batch_size, -1)
        return out
```

#### original version

```python
#@title
class LSTMnetOrigin(nn.Module):
    def __init__(self, input_dim, output_dim, hidden_dim, n_layers):
        super(LSTMnetOrigin, self).__init__()
        self.output_size = output_dim
        self.n_layers = n_layers
        self.hidden_dim = hidden_dim
        self.lstm = nn.LSTM(input_size = input_dim, 
                            hidden_size = hidden_dim, 
                            num_layers = n_layers, 
                            batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)
        self.tanh = nn.Tanh()
        self.leakyRelu = nn.LeakyReLU()
        
    def forward(self, x):
        batch_size = x.size(0)
        # x = self.tanh(x)
        lstm_out, hidden = self.lstm(x)
        lstm_out = lstm_out.contiguous().view(-1, self.hidden_dim)
        # lstm_out = self.tanh(lstm_out)
        out = self.fc(lstm_out)
        out = out.view(batch_size, -1)
        return out
```

## Function definitions for Deep Learning

In the training, the loss funcation is a weighted sum of the *prediction error* loss and *energy conservation* loss. The algorithm processes a mini batch of data at a time and use backpropagation to update model parameters. The training intends to run the specified number of epochs but will stop *early* if the returned loss increases for three consequential updates. 

For `train` and `val_test`, we include a new parameter `loss_included`, when it is assigned to `True` then we are calculating the loss with the temperature graident; otherwise, we only compute the original loss.

```python
#@title
def weighted_rmse_loss(input, target, weight):
    # defined weighted rmse loss
    # used in model training
    return torch.sqrt(torch.sum(weight * (input - target) ** 2) / torch.sum(weight))

def batch_sample_generator(size_list, batch_size):
    ## create advanced batches for training-val-test
    ## each batch should match with depth_areas for calculating ec_loss
    n_batch = int(size_list / batch_size)
    n_remain = size_list % batch_size
    batch_sampler = [list(range(i * batch_size, i * batch_size + batch_size)) for i in range(n_batch)]
    if n_remain > 0:
        batch_sampler += [list(range(n_batch * batch_size, n_batch * batch_size + n_remain))]
    return batch_sampler


def train(model, epochs, optimizer, train_loader, valid_loader, early_stopping=True, loss_included=False):
    the_last_loss = 100
    trigger_times = 0
    for i in range(epochs):
        print('This is epoch ' + str(i + 1))
        for batch, (x, y, m, p) in enumerate(train_loader):
            model.train()
            size = len(train_loader.dataset)
            x = x.to(device).float()
            y = y.to(device)
            m = m.to(device)
            p = p.to(device)
            # Compute prediction and loss
            pred = model(x)
            loss_1 = weighted_rmse_loss(pred, y, m)
            loss_2, a, b, c = calculate_ec_loss(x,
                                        pred,
                                        p,                                     
                                        depth_areas,
                                        n_depths,
                                        ec_threshold,
                                        combine_days=1)
            loss = loss_1 + elam1 * loss_2
            if loss_included:
                loss_3 = calculate_stratification(pred, n_depths, y)
                loss += elam2 * loss_3
            
            # Backpropagation
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            # Print output
            loss_track, current = loss.item(), batch * len(x)
            if batch%5 == 0:
                if loss_included:
                    print("BatLoss = " + "{:.4f}".format(loss) + \
                        ", Rc = " + "{:.4f}".format(loss_1) + \
                        ", Ec = " + "{:.4f}".format(loss_2) + \
                        ", Sc = " + "{:.4f}".format(loss_3) + \
                        f'[{current:>5d}/{size:>5d}]')
                else:
                    print("BatLoss = " + "{:.4f}".format(loss) + \
                        ", Rc = " + "{:.4f}".format(loss_1) + \
                        ", Ec = " + "{:.4f}".format(loss_2) + \
                        f'[{current:>5d}/{size:>5d}]')
        
        # Early stopping
        if early_stopping:
            the_current_loss = val_test(model, valid_loader)
            print('The current loss:', the_current_loss)
            if the_current_loss > the_last_loss:
                trigger_times += 1
                if trigger_times >= patience:
                    print('Early stopping!\nStart to test process.')
                    return model, (i+1)
            else:
                trigger_times = 0 ## reset trigger time to 0
                the_last_loss = the_current_loss

    return model, epochs
        

def val_test(model, data_loader, loss_included=False):
    # Settings
    model.eval()

    loss_total = 0
    # make predictions on validation/test data
    with torch.no_grad():
        for batch, (x, y, m, p) in enumerate(data_loader):
            x = x.to(device).float()
            y = y.to(device)
            m = m.to(device)
            p = p.to(device)

            # Compute prediction and loss
            pred = model(x)
            loss_1 = weighted_rmse_loss(pred, y, m)
            loss_2, a, b, c = calculate_ec_loss(x,
                                        pred,
                                        p,                                     
                                        depth_areas,
                                        n_depths,
                                        ec_threshold,
                                        combine_days=1)
            loss = loss_1 + elam1 * loss_2 
            if loss_included:
                loss_3 = calculate_stratification(pred, n_depths, y)
                loss += elam2 * loss_3
            loss_total += loss.item()
    return loss_total / len(data_loader)

```

```python
#@title
label_test[label_test==0] = np.nan
mask = (np.nanmean(label_test[:6,:],axis=0)-np.nanmean(label_test[35:41,:],axis=0))>5
strat_pred = label_pred[:,:,mask]
strat_test = label_test[:,mask]

mask2 = np.ndarray(strat_test.shape[1],dtype=bool)
for i in range(len(mask2)):
  if np.isnan(strat_test[:,i]).all():
    mask2[i] = False
  else:
    mask2[i] = True

strat_test_r = strat_test[:,mask2]
strat_pred_r = strat_pred[:,:,mask2]

err_mat = label_test - label_pred
strat_err = err_mat[:,:,mask]
strat_err_r = strat_err[:,:,mask2]
strat_RMSE = np.sqrt(np.nansum(strat_err_r**2, axis=1)/np.sum(~np.isnan(strat_err_r),axis=1))
strat_RMSE = strat_RMSE/np.nanstd(strat_RMSE)

strat_dT_test = np.nanmean(strat_test_r[:6,:],axis=0) - np.nanmean(strat_test_r[35:41,:],axis=0)
strat_dT_pred = np.nanmean(strat_pred_r[:,:6,:],axis=1) - np.nanmean(strat_pred_r[:,35:41,:],axis=1)
err_dT = strat_dT_pred-strat_dT_test
```

```python
#@title Temperature Sheer Error Between Models
import seaborn as sns
fig, ax = plt.subplots(1,2,figsize=(12,5))
ax[0].boxplot(np.transpose(np.abs(err_dT)))
ax[0].grid()
ax[0].set_ylabel('Temperature shear abs error')
ax[0].set_xlabel('model #')

for i in range(4):
  sns.distplot(err_dT[i,:],hist = False, kde = True,
                 kde_kws = {'linewidth': 1},
                 label = 'model '+str(i+1), ax=ax[1])

ax[1].legend()
ax[1].set_xlabel('Temperature shear error')
ax[1].grid()

plt.suptitle('Figure 9. Temperature shear error distribution', fontsize=16)
plt.show()
```

Interestingly, the models with additional loss function shows higher error with temperature shear. From the distribution, most of the models tend to predict less temperature shear compared to observation. Most of the models show long tail on the negative error side and this feature is also seen in other attempts (even shows small local peak). It is not shown here but some lake temperature profiles have gradual increase of temperature even near surface without mixing layer, but models learned the mixing layer structure usually, so it tends to stop increasing as it approaches surface. This tends to make models have bias on the temperature shear. Considering that we defined stratified lake as temperature shear over 5K, temperature shear error shows significant error for all of the models.

## Evaluation 5

Lower boundary of thermocline

Based on the temperature profile of stratified lakes (defined above), we manually set the threshold of 0.5K/m (temperature profiles are shown in appendix). We defined the deepest depth that shows gradient over 0.5K/m, which means temperature is almost equal under that depth and underlying layer can be considered as a deep water (or hypolimnion). 

```python
#@title
n = strat_pred_r.shape[2]
strat_test_grad = np.zeros((49,n))

for j in range(n):
  temp = strat_test_r[0,j]
  dz = 0
  for i in range(49):
    strat_test_grad[i,j] = (strat_test_r[i+1,j]-temp)/(0.5*(i+1-dz))
    if ~np.isnan(strat_test_r[i,j]): 
      temp = strat_test_r[i,j]
      dz = i

strat_pred_grad = (strat_pred_r[:,1:,:]-strat_pred_r[:,:-1,:])/0.5
```

```python
#@title
from scipy.interpolate import interp1d
z = np.arange(0.5,25,0.5)
d1 = np.zeros(n) # depth of thermocline lower bdy for obs
d2 = np.zeros((4,n)) # depth of thermocline lower bdy for model

for i in range(n):
  not_nan = np.logical_not(np.isnan(strat_test_grad[:,i]))
  if len(z[not_nan]) < 10:
    d1[i] = np.nan
    continue
  interp1 = interp1d(z[not_nan], strat_test_grad[not_nan,i],kind='cubic')
  z_new1 = np.arange(z[not_nan][0],z[not_nan][-1],0.05)
  test_grad_new = interp1(z_new1)
  idx_test = np.where((test_grad_new+0.5) <=0)

  if len(z_new1[idx_test]) == 0:
    d1[i] = np.nan
  else:
    d1[i] = np.max(z_new1[idx_test])

z_new2 = np.arange(z[0],z[-1],0.05)
pred_grad_new = np.zeros((4,z_new2.shape[0]))
for s in range(4):
  for i in range(n):
    interp2 = interp1d(z, strat_pred_grad[s,:,i],kind='cubic')
    pred_grad_new[s,:] = interp2(z_new2)
    idx_pred = np.where((pred_grad_new[s,:]+0.5) <=0)

    if len(z_new2[idx_pred]) == 0:
      d2[s,i] = np.nan
    else:
      d2[s,i] = np.max(z_new2[idx_pred])

err_d = d2-d1
```

```python
#@title Showing the Predicted Depth Absolute Error Between Models
mask = ~np.isnan(err_d)
filtered_data = [d[m] for d, m in zip(err_d, mask)]

# basic plot
fig, ax = plt.subplots(1,2, figsize=(12,5))
ax[0].boxplot(np.transpose(np.abs(filtered_data)))
ax[0].set_ylabel('depth abs error')
ax[0].set_xlabel('model #')
ax[0].grid()

for i in range(4):
  sns.distplot(err_d[i,:],hist = False, kde = True,
                 kde_kws = {'linewidth': 1},
                 label = 'model '+str(i+1),ax=ax[1])

ax[1].legend()
ax[1].grid()
ax[1].set_xlabel('depth error')

plt.suptitle('Figure 10. Thermocline depth error', fontsize=16)
plt.show()
```

Model 1 and 3 shows higher errors compared to 2 and 4. Relative order changes as we repeat the experiment, but model 2, which have new loss function with original structure always shows best results. Also, it shows smallest variation. Thus, selection of our new loss function to estimate structure was a good decision. Similar to temperature shear, density function shows tail on the negative side, meaning models tend to predict shallower thermocline lower boundary. For some models, small peak also shows up far from the 0 line. The lake's total depth is 25m and almost half of the profile shows error larger than 10% of the total depth.

## Introduction of the GRU Model
In addition to the LSTM model introduced in the two papers, we included a GRU (Gated Recurrent Units) model. This model was first introduced by [Cho et al (2014)](https://arxiv.org/pdf/1406.1078v3.pdf). 

There are a few major differences between an LSTM model and a GRU model. The GRU was developed to help the model use difference dependencies between time scales. We believe this could have a good impact on the Lake Mendota data because of the time-series aspect of the data. In addition, the 'gates' of the two models are different. 

> In an LSTM model, "the LSTM unit controls the amount of the new memory content being added to the memory cell independently from the forget gate. On the other hand, the GRU controls the information flow from the previous activation when computing the new, candidate activation, but does not independently control the amount of the candidate activation being added." [(Chung et al, 2014)](https://arxiv.org/pdf/1412.3555v1.pdf)


These models are similar because they both have an "additive component from t to t+1, whcih is lacking in the traditional recurrent unit" [(Chung et al, 2014)](https://arxiv.org/pdf/1412.3555v1.pdf). This additive component helps deal with the gradient vanishing problem that traditional RNN models have. 

We compare both models on the Lake Mendota data in our analysis. In the Chung et al article, they do not conclude whether the GRU model or LSTM model performs better generally. Instead, they say that it depends on the data used. However, generally, the GRU model is faster. 

For this project, where we are aiming for both speed and accuracy, we thought that trying the GRU model would be a good idea because if there are similar results and faster speed, we would prefer the GRU.
